```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path=input("Enter the file path of the result file you want to get analysis of")

df= pd.read_excel(path)

df
```

```python
grades = {"A1": 8, "A2": 7, "B1": 6, "B2": 5, "C1": 4, "C2": 3, "D1": 2, "D2": 1, "E": 0}

sheet10 = [None, 'MATHEMATICS -STANDARD', 'INFORMATION TECHNOLOGY', 'ENGLISH LANG & LIT.',
           'MATHEMATICS -BASIC', 'SCIENCE', 'SOCIAL SCIENCE', 'HINDI COURSE-A']
```

```python
choice=int(input("Enter the choice number of the corresponding subject you want the analysis of"))

if choice == 1:
    sub = "MATHEMATICS -STANDARD"
elif choice == 2:
    sub = "INFORMATION TECHNOLOGY"
elif choice == 3:
    sub = "ENGLISH LANG & LIT."
elif choice == 4:
    sub = "MATHEMATICS -BASIC"
elif choice == 5:
    sub = "SCIENCE"
elif choice == 6:
    sub = "SOCIAL SCIENCE"
elif choice == 7:
    sub = "HINDI COURSE-A"
else:
    print("Choice not found! Kindly select from given choices")
```

```python
if 'sub' in locals():
    print(f"You selected: {sub}")
```

```python
l=list(df.iloc[:,4])
t_stu=len(l)
wt=[]

for key in grades.keys():
    wt.append(l.count(key))

wtdsum = 0
n = 8
for count in wt:
    wtdsum += n * count
    n -= 1
```

```python
PI = round((wtdsum * 100 / t_stu / max(grades.values())), 2)
print("Performance Index for", sub, ":", PI)

grades_list = list(grades.keys())  
```

```python
plt.pie(wt, labels=grades_list)  # Show percentages in each slice
plt.title(f'Number of Students per Grade in {sub}')
plt.legend(title="Grades")
plt.show()
```

```python

plt.bar(grades_list, wt, color='blue', edgecolor='red')
plt.title(f'Number of Students per Grade in {sub}')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
#plt.grid()
plt.show()
```
